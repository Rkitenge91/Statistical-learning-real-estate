```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
def filter_bad_data(df, col, filters, removeNaN):
    df_c = df.copy()
    df_f = df_c
    for filter in filters:
        df_f = df_f[df_f[col]!=filter]
    if removeNaN:
        df_f = df_f[pd.isna(df_f[col]) == False]
    return df_f
```

```python
#data read in
res_df = pd.read_csv("real_estate_sales.csv")

target_col_names = ["OBJECTID", "SalePrice", "TotalAppraisedValue"]
feat_col_names = ["OBJECTID", "xrCompositeLandUseID", "xrBuildingTypeID", "xrPrimaryNeighborhoodID", "LandSF", "SaleDate", "AssrLandUse"]

res_df_filtered = filter_bad_data(res_df, "SalePrice", [0, 1], True)
target_df = res_df_filtered[target_col_names].copy()
feat_df = res_df_filtered[feat_col_names].copy()

object_id = target_df["OBJECTID"].copy()
sp = target_df["SalePrice"].copy()
tav = target_df["TotalAppraisedValue"].copy()
price_ratio = sp/tav
target_data_dict = {"ObjectID": object_id,
                    "SalePrice": sp,
                    "AppraisedValue": tav,
                    "PriceRatio": price_ratio}
target_data_df = pd.DataFrame(target_data_dict)
target_data_df.head(10)
```
